```python
from gensim.test.utils import common_texts, get_tmpfile
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
from gensim.corpora import WikiCorpus
import zhconv
import jieba
import re

cn_reg = '^[\u4e00-\u9fa5]+$'
wiki_data_raw = 'data/zhwiki-20190720-pages-articles.xml.bz2'
wiki_data_out = 'data/zhwiki.txt'
news_data_raw = 'data/sqlResult_1558435.csv'

wiki = WikiCorpus(wiki_data_raw, lemmatize=False, dictionary={})
cnt = 1
with open(wiki_data_out, 'w+', encoding='utf-8') as f:
    for text in wiki.get_texts():
        # Remove English Word
        new_text = []
        for word in text:
            if re.search(cn_reg, word):
                new_text.append(word)
        # Concat to sentence
        line =  "".join(new_text)
        # Convert to simplified Chinese
        line = zhconv.convert(line, 'zh-hans')
        # Word Segmentation
        words = jieba.cut(line)
        # Write Result
        f.write(" ".join(words) + '\n')
        if cnt % 1000 == 0:
            print("Handling %d texts..." % cnt)
        cnt+=1
        
```

```python
model = Word2Vec.load("word2vec.model")
model.train([["hello", "world"]], total_examples=1, epochs=1)
```
